Join the data from Part 1 with the data from Part 2 to create a new dataframe.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('clean_vancouver_bike_data_df.csv')

df2 = pd.read_csv('van_rad_data_df.csv')
```

```python
fsq_van_bikes_df = pd.merge(df1, df2, on="lat_long")
fsq_van_bikes_df = fsq_van_bikes_df.drop(['latitude','longitude'],axis=1)
fsq_van_bikes_df
```

```python
fsq_van_bikes_df.to_csv('fsq_van_bikes_df.csv', index=False)
```

# Database

Put all your results in an SQLite3 database (remember, SQLite stores its databases as files in your local machine - make sure to create your database in your project's data/ directory!)

```python
import sqlite3
from sqlite3 import Error

def create_connection(path):
    connection = None
    try:
        connection = sqlite3.connect(path)
        print("Connection to SQLite DB successful")
    except Error as e:
        print(f"The error '{e}' occurred")

    return connection
```

```python
connection = create_connection("fsq_van_bikes_df.sqlite")
```

```python
conn = sqlite3.connect("fsq_van_bikes_df.sqlite")
fsq_van_bikes_df.to_sql("fsq_van_bike_data", conn, if_exists='replace')
```

```python
# all data from db
df_all = pd.read_sql("""SELECT 
                            *
                        FROM fsq_van_bike_data
                     """, conn
                    )
df_all
```
