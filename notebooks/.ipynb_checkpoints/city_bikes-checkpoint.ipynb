Put your parsed results into a DataFrame.

```python
city_bike_url = "https://api.citybik.es/v2/networks/mobibikes"

response = requests.request("GET",city_bike_url)
data = response.json()
stations = data['network']['stations']
van_station_df = pd.json_normalize(data = stations)
print(van_station_df.shape)
print(van_station_df['free_bikes'].sum())
print(van_station_df['empty_slots'].sum())
van_station_df
```

```python
van_station_df.dropna(subset=['latitude', 'longitude'], inplace=True)
van_station_df['lat_long'] = van_station_df['latitude'].astype(str) + ',' + van_station_df['longitude'].astype(str)
van_station_df = van_station_df[van_station_df['lat_long'] != '0.0,0.0']
van_station_df
```

```python
locations = van_station_df.lat_long.tolist()
#locations    
```

```python
van_station_df['total_bikes'] = van_station_df['empty_slots'] + van_station_df['free_bikes']
clean_van_df = van_station_df
clean_van_df = clean_van_df.drop(['empty_slots', 'free_bikes', 'id','timestamp','extra.ebikes','extra.has_ebikes', 'extra.last_updated',
                                   'extra.normal_bikes','extra.renting','extra.returning','extra.slots','extra.uid'],axis=1)
clean_van_df
```

```python
clean_van_df.to_csv('clean_vancouver_bike_data_df.csv')
```
