Get the top 10 restaurants according to their rating

```python
yelp_url = "https://api.yelp.com/v3/businesses/search"
yelp_headers = {
    'Accept': "application/json",
    "Authorization": "Bearer " + yelp_api_key
}

yelp_params = {
    "latitude": '49.262487',
    "longitude": '-123.114397',
    "categories":'bar',
    "radius":'5000',
    "limit":'10',
    "sort_by":'rating'
}

response = requests.request("GET",yelp_url, params=yelp_params,headers=yelp_headers)
data = response.json()

yelp_top10_df = pd.json_normalize(data = data['businesses'])
yelp_top10_df
```
